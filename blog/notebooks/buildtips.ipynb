```python
feature.on('keys', e => {  
  k = e.keys
  player = parcel.players[0]
  boat = parcel.getFeatureById('boat') 
  x = boat.position.x
  y = boat.position.y
  z = boat.position.z 
  player.position.x = player.position.x + 1 
  //if(k.a){ y = y + 1 }
  //else if(k.b){ y = y - 1 }
  if(k.down){ x = x + 1 }
  else if(k.up){ x = x - 1 }
  else if(k.right){ z = z + 1 }
  else if(k.left){ z = z - 1 }
  boat.set({'position': [x, y, z] }) 
} )
```
